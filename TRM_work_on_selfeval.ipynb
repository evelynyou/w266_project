### Cull based on selfeval
Remove any dialogues from train, valid, test that did not have '5' for both participants.  The first rating is empathy (others are relevance and understandability).  Then save all three so that BERT can be finetuned on revised ED set.


```python
data_dir='/home/tmetz/w266_project/data/empatheticdialogues'
file_name='train.csv'
ed_train=pd.read_csv(os.path.join(data_dir, file_name), error_bad_lines=False,quoting=csv.QUOTE_NONE)
print(ed_train.shape)
ed_train.head(20)
```

```python
ed_train.selfeval.head(20)
```

```python
ed_train['empathy1']=ed_train.selfeval.str[0]
ed_train['empathy2']=ed_train.selfeval.str[6]
ed_train['combined_empathy']=ed_train.empathy1+ed_train.empathy2
print (ed_train.shape)
ed_train.head(5)
```

```python
ed_train.empathy1.value_counts()
```

```python
#keep only those conversations where both sides rate empathy as 5
ed_train_revised=ed_train[(ed_train.empathy1=='5')&(ed_train.empathy2=='5')]
print (ed_train_revised.shape)
```

```python
#now save back out as a csv file
data_dir='/home/tmetz/w266_project/data/ed_revised_selfeval'
file_name='train.csv'
ed_train_revised.to_csv(os.path.join(data_dir,file_name),quoting=csv.QUOTE_NONE,index=False)
```

```python
ed_train.selfeval.str[6].head(20)
```

```python
#now validation set
data_dir='/home/tmetz/w266_project/data/empatheticdialogues'
file_name='valid.csv'
ed=pd.read_csv(os.path.join(data_dir, file_name), error_bad_lines=False,quoting=csv.QUOTE_NONE)
ed['empathy1']=ed.selfeval.str[0]
ed['empathy2']=ed.selfeval.str[6]
ed['combined_empathy']=ed.empathy1+ed.empathy2
ed_revised=ed[(ed.empathy1=='5')&(ed.empathy2=='5')]
data_dir='/home/tmetz/w266_project/data/ed_revised_selfeval'
file_name='valid.csv'
ed_revised.to_csv(os.path.join(data_dir,file_name),quoting=csv.QUOTE_NONE,index=False)
```

```python
#now test set
#now validation set
data_dir='/home/tmetz/w266_project/data/empatheticdialogues'
file_name='test.csv'
ed=pd.read_csv(os.path.join(data_dir, file_name), error_bad_lines=False,quoting=csv.QUOTE_NONE)
ed['empathy1']=ed.selfeval.str[0]
ed['empathy2']=ed.selfeval.str[6]
ed['combined_empathy']=ed.empathy1+ed.empathy2
ed_revised=ed[(ed.empathy1=='5')&(ed.empathy2=='5')]
data_dir='/home/tmetz/w266_project/data/ed_revised_selfeval'
file_name='test.csv'
ed_revised.to_csv(os.path.join(data_dir,file_name),quoting=csv.QUOTE_NONE,index=False)
```
